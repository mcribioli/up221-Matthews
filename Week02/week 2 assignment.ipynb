```python
###Assignment 1 Week 2
#In this assigment, I will be conducting a preliminary analysis on the California 2019-2014 5 year ACS survey

import pandas as pd
import geopandas as gpd

#read census data for california
census = pd.read_csv('Data/Comprehensive Report of ACS 2019-2014 California Census Tracts.csv')
```

```python
#Sort the census tracts by median age to find the 5 youngest tracts in california
census_sorted = census.sort_values(by=['SE_A01004_001'])
census_sorted=census_sorted.head()
```

```python
#Keep only the columns of census tract and median age
census_sorted=census_sorted[["Geo_QName","SE_A01004_001"]]
#rename the columns
census_sorted.columns = ['Census_Tract','Median_Age']
```

```python
#Plot the results
ax = census_sorted.plot.bar(x='Census_Tract',y='Median_Age',title = 'Lowest 5 Median Age Census Tracts in California')
```
